```python
# import numpy as np
import torch
import matplotlib.pyplot as plt
import math
```

```python
x = torch.linspace(-math.pi, math.pi,2000)
y = -34.7 + 12.9*x + 8.1145 * x ** 2 + 8 * x ** 3
print(x.shape, y.shape)

plt.plot(x, y, 'r' ,label="Data Points")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
class CurveFitting():
  def __init__(self):
    self.a = None
    self.b = None
    self.c = None
    self.d = None
    self.lr = None
    # self.d = None

  def firCurve(self, x,y):
    epoch = 1000
    # self.a = -30
    # self.b = 12
    # self.c = 8
    # self.d = 6
    self.a = torch.randn(())
    self.b = torch.randn(())
    self.c = torch.randn(())
    self.d = torch.randn(())
    self.lr = 0.000001
    print("Before Weight", self.a, self.b, self.c , self.d)
    for i in range(epoch):
      y_pred = self.a + self.b * x + self.c * x ** 2 + self.d * x ** 3
      grad_y_pred = 2 * (y_pred - y)
      # grad_y_pred = grad_y_pred.sum()
      # grad_a = np.sum(grad_y_pred)
      # grad_b = np.sum(grad_y_pred * x)
      # grad_c = np.sum(grad_y_pred * x ** 2)
      # grad_d = np.sum(grad_y_pred * x ** 3)
      grad_a = grad_y_pred.sum()
      # print("1", grad_a)
      grad_b = (grad_y_pred * x).sum()
      grad_c = (grad_y_pred * x ** 2).sum()
      grad_d = (grad_y_pred * x ** 3).sum()
      # print(grad_a, grad_b, grad_c, grad_d)
      self.a -= self.lr * grad_a
      self.b -= self.lr * grad_b
      self.c -= self.lr * grad_c
      self.d -= self.lr * grad_d
      # return self.a, self.b, self.c, self.d
      # return
    print("After Weight", self.a, self.b, self.c , self.d)

    return

  def predict(self, x):
    y_pred = self.a + self.b * x + self.c * x ** 2 + self.d * x ** 3
    # print(y_pred)
    return y_pred

  # def loss(x):


```

```python
m = CurveFitting()
m.firCurve(x,y)
```

```python
m.predict(x)
```

```python
plt.plot(x, m.predict(x), 'r' ,label="Predicted")
plt.plot(x, y, 'b' ,label="Data Points")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
