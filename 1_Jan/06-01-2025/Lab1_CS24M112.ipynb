```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-math.pi, math.pi,2000).reshape(-1,1)
y = -34.7 + 12.9*x + 8.1145 * x ** 2 + 8 * x ** 3
y = y.reshape(-1,1)
print(x.shape, y.shape)

plt.plot(x, y, 'r' ,label="Data Points")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
class CurveFitting():
  def __init__(self):
    self.w0 = None
    self.w1 = None
    self.w2 = None
    self.w3 = None
    self.lr = None

  def firCurve(self, x,y):
    epoch = 10000
    self.w0 = np.random.randint(0,5)
    self.w1 = np.random.randint(0,5)
    self.w2 = np.random.randint(0,5)
    self.w3 = np.random.randint(0,5)
    self.lr = 0.0000001
    print("Before Grad", self.w0, self.w1, self.w2 , self.w3)
    for i in range(epoch):
      y_pred = self.w0 + self.w1*x + self.w2 * x ** 2 + self.w3 * x ** 3
      grad_y_pred = 2 * (y_pred - y)
      grad_a = np.sum(grad_y_pred)
      grad_b = np.sum(grad_y_pred * x)
      grad_c = np.sum(grad_y_pred * x ** 2)
      grad_d = np.sum(grad_y_pred * x ** 3)
      self.w0 -= self.lr * grad_a
      self.w1 -= self.lr * grad_b
      self.w2 -= self.lr * grad_c
      self.w3 -= self.lr * grad_d
    print("After Grad", self.w0, self.w1, self.w2 , self.w3)
    return self.w0, self.w1, self.w2, self.w3

  def predict(self, x):
    # print(x.shape)
    y_pred = self.w0 + self.w1 * x + self.w2 * x ** 2 + self.w3 * x ** 3
    # y_pred = -88 + 12.9*x + 8.1145 * x ** 2 + 8 * x ** 3
    return y_pred

  def loss(self, y):
    # print(y.shape)
    return np.mean((self.predict(y) - y)**2) ** (1/2)


```

```python
m = CurveFitting()
m.firCurve(x,y)
```

```python
plt.plot(x, m.predict(x), 'r' ,label="Predicted")
plt.plot(x, y, 'b' ,label="Data Points")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
