# Laboratory Exercise Tasks: Seaborn Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
diamonds = sns.load_dataset("diamonds")
penguins = sns.load_dataset("penguins").dropna()
```

# Excercise 1

```python
# ========== Exercise 1 ==========
# 1. Histogram + KDE
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Histogram of Total Bill with KDE")
plt.xlabel("Total Bill")
plt.ylabel("Count")
plt.show()

# 2. Stacked histogram by sex
sns.histplot(data=tips, x="total_bill", hue="sex", multiple="stack")
plt.title("Stacked Histogram of Total Bill by Sex")
plt.xlabel("Total Bill")
plt.ylabel("Count")
plt.show()

# 3. ECDF plot
sns.ecdfplot(data=tips, x="tip")
plt.title("ECDF of Tips")
plt.xlabel("Tip")
plt.ylabel("ECDF")
plt.show()
```

# Excersise 2

```python
# 1. Box plot of total_bill by day, hue=time
sns.boxplot(data=tips, x="day", y="total_bill", hue="time", palette="Set1")
plt.title("Box Plot of Total Bill by Day and Time")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.legend(title="Time")
plt.show()

# 2. Violin plot for iris
sns.violinplot(data=iris, x="species", y="petal_length", palette="muted")
plt.title("Violin Plot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.show()

# 3. Count plot of day values grouped by smoker
sns.countplot(data=tips, x="day", hue="smoker", palette="pastel")
plt.title("Count of Days Grouped by Smoker")
plt.xlabel("Day")
plt.ylabel("Count")
plt.legend(title="Smoker")
plt.show()
```

# Excercise 3

```python
# ========== Exercise 3 ==========
# 1. Scatter plot of total_bill vs tip
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", palette="Set2")
plt.title("Total Bill vs Tip by Day")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend(title="Day")
plt.show()

# 2. Regression line with lmplot by smoker
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", aspect=1.5)
plt.title("Regression: Total Bill vs Tip by Smoker")
plt.show()

# 3. Diamonds scatter with annotation
plt.figure(figsize=(8, 6))
sns.scatterplot(data=diamonds.sample(1000), x="carat", y="price", alpha=0.6)
plt.title("Carat vs Price in Diamonds")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.axhline(10000, color="red", linestyle="--")
plt.text(0.3, 10500, "Expensive", color="red")
plt.axhline(3000, color="green", linestyle="--")
plt.text(0.3, 3200, "Affordable", color="green")
plt.show()
```

# Excercise 4

```python
# ========== Exercise 4 ==========
# 1. Correlation heatmap for tips
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Tips Dataset")
plt.show()

# 2. Pair plot for iris
sns.pairplot(iris, hue="species", kind="kde", diag_kind="kde", palette="husl")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()

# 3. Joint plot sepal_length vs sepal_width
sns.jointplot(data=iris, x="sepal_length", y="sepal_width", kind="scatter")
plt.suptitle("Joint Plot: Sepal Length vs Width", y=1.02)
plt.show()
```

# Excercise 5

```python
# ========== Exercise 5 ==========
# 1. FacetGrid histograms by day
g = sns.FacetGrid(tips, col="day")
g.map(sns.histplot, "total_bill")
plt.suptitle("Facet Histograms of Total Bill by Day", y=1.05)
plt.show()

# 2. Displot with time and hue by sex
sns.displot(data=tips, x="total_bill", col="time", hue="sex", multiple="stack")
plt.show()
```

# Excercise 6

```python
# ========== Exercise 6 ==========
# 1. Swarm plot of flipper length by species
sns.swarmplot(data=penguins, x="species", y="flipper_length_mm", palette="Set3")
plt.title("Swarm Plot of Flipper Length by Species")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.show()

# 2. Boxen plot of body mass by species
sns.boxenplot(data=penguins, x="species", y="body_mass_g", palette="cool")
plt.title("Boxen Plot of Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
```
